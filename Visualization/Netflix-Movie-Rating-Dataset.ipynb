# II. Tìm hiểu và tiền xử lý dữ liệu 🛠️

### 1. Các thư viện sử dụng 📚

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. Đọc dữ liệu 📂

```python
# Đọc dữ liệu từ file CSV
rating_data = pd.read_csv('Netflix_Dataset_Rating.csv')
movie_data = pd.read_csv('Netflix_Dataset_Movie.csv')
```

### 3. Các thông tin chung về dữ liệu 📝

##### 3.1. Đếm số dòng và số cột

```python
num_rows, num_cols = rating_data.shape
print("Số dòng của bảng dữ liệu Rating:", num_rows)
print("Số cột của bảng dữ liệu Rating:", num_cols)
```

```python
num_rows, num_cols = movie_data.shape
print("Số dòng của bảng dữ liệu Movie:", num_rows)
print("Số cột của bảng dữ liệu Movie:", num_cols)
```

##### 3.2. Viết bảng mô tả về các cột

Mô tả về dữ liệu của bảng Movie:
|Cột|Mô tả|
|---|---|
|Movie_ID|	Định danh duy nhất cho mỗi bộ phim|
|Year|	Năm phát hành của bộ phim|
|Name|	Tên của bộ phim|

Mô tả về dữ liệu của bảng Rating:
|Cột|Mô tả|
|---|---|
|User_ID|	Định danh duy nhất cho mỗi người dùng|
|Rating|	Điểm đánh giá được người dùng đưa ra cho bộ phim|
|Movie_ID|	ID của bộ phim được người dùng đánh giá|

### 4. Tiền xử lý dữ liệu 🧹

##### 4.1. Gộp 2 tập dữ liệu thành một tập duy nhất

Kết hợp dữ liệu từ hai dataframe theo cột Movie_ID

```python
combined_data = pd.merge(rating_data, movie_data, on='Movie_ID')
```

Tính trung bình đánh giá và số lượng đánh giá cho mỗi bộ phim

```python
combined_data = combined_data.groupby(['Movie_ID', 'Name', 'Year']).agg({'Rating': 'mean', 'User_ID': 'count'}).reset_index()
```

Đổi tên các cột User_ID và Rating

```python
# Đổi tên cột User_ID thành Count
combined_data = combined_data.rename(columns={'User_ID': 'Count'})

# Đổi tên cột Rating thành Average_Rating
combined_data = combined_data.rename(columns={'Rating': 'Average_Rating'})
```

##### 4.2. Kiểm tra số dòng bị trùng lặp và các cột có giá trị thiếu

Kiểm tra số dòng bị trùng lặp

```python
print(f"Dữ liệu có {combined_data.duplicated().sum()} dòng bị trùng.")
```

Kiểm tra các cột có giá trị thiếu

```python
combined_data.isnull().sum().sort_values(ascending=False)
```

```python
combined_data.head(5)
```

# III. Phân tích dữ liệu 🔍

Để khám phá và rút ra được các kết luận từ dữ liệu, ta sẽ thực hiện trả lời các câu hỏi để từ đó đưa ra các phân tích và trực quan hóa dữ liệu.

### 1. Mỗi User thường tham gia đánh giá bao nhiêu bộ phim?

Đếm số lượng rating mà mỗi User ID thực hiện và đặt tên cột mới là 'Rating Per User Count'

```python
user_rating_count = rating_data['User_ID'].value_counts().reset_index()

user_rating_count.columns = ['User_ID', 'Rating Per User Count']
```

Tạo ra mô tả cho cột Rating Per User Count

```python
user_rating_count['Rating Per User Count'].describe()
```

Tạo biểu đồ density plot cho cột Rating Per User Count

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(user_rating_count['Rating Per User Count'], color='skyblue', fill=True)
plt.title('Biểu đồ mật độ của `Rating Per User Count`')
plt.xlabel('Rating Per User Count')
plt.ylabel('Density')

# Tính các giá trị thống kê
mean_value = user_rating_count['Rating Per User Count'].mean()
median_value = user_rating_count['Rating Per User Count'].median()
percentile_25 = user_rating_count['Rating Per User Count'].quantile(0.25)
percentile_75 = user_rating_count['Rating Per User Count'].quantile(0.75)
min_value = user_rating_count['Rating Per User Count'].min()
max_value = user_rating_count['Rating Per User Count'].max()

# Vẽ các đường thẳng đứng cho giá trị trung bình, trung vị và phân vị
plt.axvline(mean_value, color='red', linestyle='--', label='Mean')
plt.axvline(median_value, color='yellow', linestyle='--', label='Median')
plt.axvline(percentile_25, color='green', linestyle='--', label='25th Percentile')
plt.axvline(percentile_75, color='blue', linestyle='--', label='75th Percentile')

# Vẽ các đường thẳng đứng cho giá trị min và max
plt.axvline(min_value, color='black', linestyle='--', label='Min')
plt.axvline(max_value, color='black', linestyle='--', label='Max')

plt.legend()
plt.show()
```

- Phân phối của số bộ phim mỗi người đánh giá (Rating Per User Count) bị lệch phải, có miền giá trị từ 5 đến 1344, với giá trị trung bình là 120.85 và độ lệch chuẩn là 79.78. 

- Phân phối này cũng cho thấy rằng có một số người đánh giá rất nhiều bộ phim, với một số người đánh giá lên đến 1344 bộ phim.

=> Điều này cho thấy rằng phân phối của số lượng bộ phim mỗi người đánh giá không đồng đều, với một số người đánh giá ít bộ phim hơn so với số người đánh giá nhiều bộ phim. Điều này có thể là do một số người dùng thích xem nhiều bộ phim hơn so với số người dùng khác hoặc họ là những người làm công việc liên quan đến việc review phim.

### 3. Mỗi mức Rating chiếm tỉ lệ bao nhiêu? Mức Rating trung bình của mọi bộ phim là bao nhiêu?

Đếm số lượng rating cho mỗi loại rating

```python
rating_counts = rating_data['Rating'].value_counts().sort_index(ascending=False)

# Tạo labels cho các phần của pie chart
labels = ['Rating 5', 'Rating 4', 'Rating 3', 'Rating 2', 'Rating 1']
```

Tạo biểu đồ pie chart cho số lượt rating của mỗi loại rating

```python
plt.figure(figsize=(5, 5))
plt.pie(rating_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'], wedgeprops=dict(width=0.3))
plt.title('Tỉ lệ phần trăm số lượt rating cho mỗi loại rating', fontsize=14)

# Thêm tổng số lượt rating và trung bình giá trị rating vào giữa donut chart
# Tính tổng số lượt rating
total_ratings = len(rating_data)

# Tính trung bình của giá trị rating
average_rating = rating_data['Rating'].mean()

# Thêm thông tin về tổng số lượt rating vào giữa donut
plt.text(0, 0.1, f'Total: {total_ratings}', fontsize=12, color='black', ha='center')

# Thêm thông tin về trung bình giá trị rating vào giữa donut
plt.text(0, -0.1, f'Mean: {average_rating:.2f}', fontsize=12, color='black', ha='center')

# Hiển thị biểu đồ
plt.axis('equal')  # Đảm bảo biểu đồ là hình tròn
plt.tight_layout()
plt.show()
```

- Tỉ lệ phần trăm số lượt rating cho mỗi loại rating theo thứ tự giảm dần là: Rating 4 (34.0%) > Rating 3 (29.9%) > Rating 5 (21.7%) > Rating 2 (10.1%) > Rating 1 (4.2%).
Điều này cho thấy người dùng thường có xu hướng đánh giá cao cho các bộ phim mà họ xem. 
- Tổng số lượt rating là 17,337,458 và giá trị trung bình của rating là 3.59.
Tỉ lệ phần trăm số lượt rating cho mỗi loại rating phân bố không đồng đều, có xu hướng chấm điểm cao hơn mức trung bình là 3 sao.

### 4. Cho biết Top 10 bộ phim có tổng giá trị điểm rating cao nhất với điều kiện có trên 3000 lượt đánh giá và rating cao hơn mức trung bình là những bộ phim nào?

Tìm ra Top 10 bộ phim có tổng giá trị điểm rating cao nhất

```python
# Lọc ra các bộ phim có số lượng đánh giá lớn hơn 3000 và điểm rating trung bình lớn hơn 3.59
clone_combined_data = combined_data[(combined_data['Count'] > 3000) & (combined_data['Average_Rating'] > 3.59)]

# Tính tổng giá trị điểm rating cho mỗi bộ phim
combined_data['Total_Rating_Value'] = combined_data['Count'] * combined_data['Average_Rating']

# Sắp xếp các bộ phim theo tổng giá trị điểm rating từ cao đến thấp
top_total_rating_value = combined_data.nlargest(10, 'Total_Rating_Value')

# In ra 10 bộ phim có tổng giá trị điểm rating cao nhất
top_total_rating_value
```

Tính tổng giá trị điểm rating trung bình của tất cả các bộ phim

```python
average_rating_value_all_movies = combined_data['Total_Rating_Value'].sum() / len(combined_data)
```

Vẽ biểu đồ cột cho top 10 bộ phim có tổng giá trị điểm rating cao nhất

```python
top_total_rating_value = top_total_rating_value[::-1]

plt.figure(figsize=(10, 6))
plt.barh(top_total_rating_value['Name'], top_total_rating_value['Total_Rating_Value'], color='skyblue')
plt.axvline(x=average_rating_value_all_movies, color='teal', linestyle='--', linewidth=1, label='Average of Total Rating Value of All Movies')
plt.title('Top 10 bộ phim có tổng giá trị điểm rating cao nhất', fontsize=14)
plt.xlabel('Total Rating Value')
plt.ylabel('Movie')
plt.legend()
plt.tight_layout()
plt.show()
```

- Bộ phim có tổng giá trị điểm rating cao nhất là "Pirate of the Caribbean: The Curse of the Black Pearl" với gần 500000 điểm rating, theo sau là "The Lord of the Rings: The Fellowship of the Ring", "The Sixth Sense", "Finding Nemo", "The Silence of the Lambs" với hơn 400000 điểm rating.

- Những bộ phim còn lại trong Top 10 cũng có tổng giá trị điểm rating khá cao, từ 338000 đến 400000 điểm rating.

=> Những bộ phim lọt vào trong danh sách này có thể được coi là những bộ phim có sức hút cực kì lớn đối với người xem so với các bộ phim khác trong tập dữ liệu bởi vì nó vừa có số lượng đánh giá lớn, vừa có điểm rating trung bình cao với công thức xếp hạng là tổng giá trị điểm rating.

### 10. Cho biết Top 5 series phim có Rating trung bình cao nhất? Các phần trong Series đó nhận được mức đánh giá nào? 

Tạo ra danh sách 5 Series phim có Rating trung bình cao nhất

```python
# Lọc các bộ phim có kí tự ':' trong tên phim
season_movies = pd.DataFrame(combined_data[combined_data['Name'].str.contains(':', case=False, regex=False)])

# Tạo cột 'Average_Rating' từ dữ liệu rating đã có
season_movies['Series'] = season_movies['Name'].str.extract(r'^([^:]+):', expand=False)

# Gom nhóm các bộ phim theo tên của series và tính rating trung bình của mỗi series
grouped_season_movies = season_movies.groupby('Series')['Average_Rating'].mean().reset_index()

# Tính số lượng phần phim cho mỗi series
series_counts = season_movies.groupby('Series').size().reset_index(name='Number_of_Movies')

# Tính trung bình số lượt đánh giá cho mỗi series
average_counts = season_movies.groupby('Series')['Count'].mean().reset_index(name='Avg_Count').round(0)

# Gộp cả hai DataFrame với DataFrame đã có
grouped_season_movies = pd.merge(grouped_season_movies, series_counts, on='Series')
grouped_season_movies = pd.merge(grouped_season_movies, average_counts, on='Series')

# Lọc ra các series có ít nhất 4 phim
grouped_season_movies = grouped_season_movies[grouped_season_movies['Number_of_Movies'] >= 4]

# Sắp xếp theo cột 'Average_Rating' theo thứ tự giảm dần và reset index
sorted_grouped_season_movies = grouped_season_movies.sort_values(by='Average_Rating', ascending=False).reset_index(drop=True)

sorted_grouped_season_movies
```

Tạo biểu đồ thể hiện mức Rating qua các phần của từng Serie phim

```python
# Danh sách 5 series đạt được rating trung bình cao nhất
top_series = sorted_grouped_season_movies['Series'].head(5)

# Lọc ra các bộ phim trong 5 series đó từ DataFrame combined_data
top_series_movies = combined_data[combined_data['Name'].str.contains('|'.join(top_series))].copy()

# Thêm cột 'Series_Name' vào DataFrame top_series_movies
top_series_movies['Series_Name'] = top_series_movies['Name'].str.extract(r'^([^:]+):', expand=False)

# Màu sắc cho mỗi series
colors = ['b', 'g', 'r', 'orange', 'm']

# Vẽ đồ thị cho từng series
fig, ax = plt.subplots(figsize=(12, 8))

# Lưu lại các giá trị hoành độ của các đường kẻ lưới dọc
grid_lines_x = []

# Tính toán vị trí của các đường kẻ lưới dọc
for i in range(len(top_series) + 1):
    grid_line_x = i - 0.5
    ax.axvline(x=grid_line_x, color='black', linewidth=0.4)
    grid_lines_x.append(grid_line_x)

# Danh sách các giá trị hoành độ tương ứng cho 5 đường trung bình
average_line_x = [grid_lines_x[i] for i in range(6)]

for i, series in enumerate(top_series):
    # Lấy dữ liệu của từng series
    series_data = top_series_movies[top_series_movies['Name'].str.contains(series)]
    
    # Tính rating trung bình cho mỗi bộ phim
    avg_ratings = series_data.groupby(['Name', 'Year'])['Average_Rating'].mean()

    # Sắp xếp các bộ phim trong cùng 1 series theo năm sản xuất
    avg_ratings = avg_ratings.sort_index(level='Year')

    # Tạo các điểm cho mỗi bộ phim
    x_values = np.linspace(i - 0.4, i + 0.4, len(avg_ratings))  # Mỗi bộ phim chiếm 1/5 trục Ox
    ax.scatter(x_values, avg_ratings.values, label=series, color=colors[i])

    # Nối các điểm với nhau để tạo thành đường
    for j in range(len(avg_ratings) - 1):
        ax.plot([x_values[j], x_values[j+1]], [avg_ratings.values[j], avg_ratings.values[j+1]], color=colors[i])
    
# Đặt tên cho các vùng trên trục Ox
ax.set_xticks(np.arange(len(top_series)))
ax.set_xticklabels(top_series)

# Xóa dấu gạch ngang ở mỗi tên series
ax.tick_params(axis='x', length=0)

# Đặt tiêu đề và nhãn cho trục y
ax.set_title('Top 5 series phim có rating trung bình cao nhất', fontsize=16)
ax.set_ylabel('Average Rating')
ax.legend()

# Tính toán vị trí của các đường kẻ lưới ngang 
for rating in np.arange(3.4, 4.7, 0.1):
    ax.axhline(y=rating, color='gray', linestyle='--', linewidth=0.4)


# Hiển thị đồ thị
plt.show()
```

- Serie The Simpsons là Serie thành công nhất về cả mặt Rating trung bình và số lượng đánh giá trung bình nhận được. Cả 5 phần phim đều có mức Rating ổn định từ 4.3 đến 4.6.
- Serie Stargate SG-1 cũng có mức Rating trung bình cao không kém cạnh Serie The Simpsons với mức Rating trung bình chỉ thấp hơn 0.07 điểm. Cả 4 phần phim trong Serie này đều có mức Rating 4.2 đến 4.5.

Mức Rating qua các phần có biến động lớn với 2 phần dưới mức 4.0. Trong đó có một phần chỉ đạt mức Rating 3.4, tức là dưới ngưỡng trung bình (3.59) một khoảng khá lớn.

### 11. Các phần của Top 5 Series phim đạt được rating cao nhất có mức Rating trung bình ở mức nào so với mức Rating trung bình của mọi bộ phim tại năm phần phim đó được sản xuất?

```python
# Danh sách 5 series thành công nhất
top_series = sorted_grouped_season_movies['Series'].head(5)

# Lọc ra các bộ phim trong 5 series đó từ DataFrame combined_data
top_series_movies = combined_data[combined_data['Name'].str.contains('|'.join(top_series))].copy()

# Thêm cột 'Series_Name' vào DataFrame top_series_movies
top_series_movies.loc[:, 'Series_Name'] = top_series_movies['Name'].str.extract(r'^([^:]+):', expand=False)

# Gom nhóm các phim theo năm và tính trung bình rating cho mỗi năm
average_rating_per_year = combined_data.groupby('Year')['Average_Rating'].mean()

# Lọc dữ liệu từ năm 1960 trở đi
filtered_data = average_rating_per_year.loc[1960:]

colors = ['b', 'g', 'r', 'orange', 'm']

# Vẽ đồ thị cho từng series và trung bình rating qua các năm
fig, ax = plt.subplots(figsize=(12, 8))

for i, series in enumerate(top_series):
    # Lấy dữ liệu của từng series
    series_data = top_series_movies[top_series_movies['Name'].str.contains(series)]

    # Groupby năm và tính trung bình rating cho mỗi năm
    yearly_avg_ratings = series_data.groupby('Year')['Average_Rating'].mean()
    
    # Vẽ đồ thị cho từng series
    ax.plot(yearly_avg_ratings.index, yearly_avg_ratings.values, label=series, color=colors[i])

    # Thêm chấm cho mỗi tập phim
    for year, rating in zip(yearly_avg_ratings.index, yearly_avg_ratings.values):
        ax.scatter(year, rating, color=colors[i])

# Vẽ đồ thị cho trung bình rating qua các năm
ax.plot(filtered_data.index, filtered_data.values, label='Average rating', color='teal', linestyle='--')

# Đặt tiêu đề và nhãn cho trục x và trục y
ax.set_title('Top 5 Series phim đạt được rating cao nhất so với mức rating trung bình qua các năm') 
ax.set_xlabel('Year')
ax.set_ylabel('Average Rating')
ax.legend()

for rating in np.arange(3.4, 4.7, 0.1):
    ax.axhline(y=rating, color='gray', linestyle='--', linewidth=0.4)

for year in np.arange(1960, 2006, 5):
    ax.axvline(x=year, color='black', linewidth=0.4)
    
plt.show()
```

- Đa phần các phần phim trong Top 5 series đều có rating trung bình cao hơn so với mức rating trung bình tất cả các phim của năm đó, ngoại trừ một số phần phim của series The Twilight Zone vào năm 1963 và một số phần phim của series Star Trek vào năm 1979.

- The Simpsons và Stargate SG-1 là hai series có mọi phần đều đạt mức Rating vượt trội so với ngưỡng trung bình của năm mà các phần phim đó được sản xuất.

- Geogre Carlin là Serie có mức rating thấp nhất trong Top 5 series, nhưng mọi phần phim của series này vẫn đạt mức rating trung bình cao hơn so với mức rating trung bình trong năm mà các phần của nó được sản xuất.

- Các Series trong Top 5 đều có phần lớn các phần phim nằm trong giai đoạn từ năm 1990 trở đi, chỉ có riêng The Twilight Zone là toàn bộ các phần được sản xuất trong giai đoạn 1960-1968.

=> Các series phim trong Top 5 đều là những series phim thành công và được đánh giá cao trong mắt khán giả. Với 4/5 Series phim có các phần nằm trong giai đoạn từ năm 1990 trở đi, cho thấy dòng phim nhiều tập ngày càng phát triển mạnh mẽ và trở nên phổ biến hơn.
