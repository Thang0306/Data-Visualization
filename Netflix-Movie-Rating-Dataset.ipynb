# II. T√¨m hi·ªÉu v√† ti·ªÅn x·ª≠ l√Ω d·ªØ li·ªáu üõ†Ô∏è

### 1. C√°c th∆∞ vi·ªán s·ª≠ d·ª•ng üìö

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. ƒê·ªçc d·ªØ li·ªáu üìÇ

```python
# ƒê·ªçc d·ªØ li·ªáu t·ª´ file CSV
rating_data = pd.read_csv('Netflix_Dataset_Rating.csv')
movie_data = pd.read_csv('Netflix_Dataset_Movie.csv')
```

### 3. C√°c th√¥ng tin chung v·ªÅ d·ªØ li·ªáu üìù

##### 3.1. ƒê·∫øm s·ªë d√≤ng v√† s·ªë c·ªôt

```python
num_rows, num_cols = rating_data.shape
print("S·ªë d√≤ng c·ªßa b·∫£ng d·ªØ li·ªáu Rating:", num_rows)
print("S·ªë c·ªôt c·ªßa b·∫£ng d·ªØ li·ªáu Rating:", num_cols)
```

```python
num_rows, num_cols = movie_data.shape
print("S·ªë d√≤ng c·ªßa b·∫£ng d·ªØ li·ªáu Movie:", num_rows)
print("S·ªë c·ªôt c·ªßa b·∫£ng d·ªØ li·ªáu Movie:", num_cols)
```

##### 3.2. Vi·∫øt b·∫£ng m√¥ t·∫£ v·ªÅ c√°c c·ªôt

M√¥ t·∫£ v·ªÅ d·ªØ li·ªáu c·ªßa b·∫£ng Movie:

|Movie_ID|	ƒê·ªãnh danh duy nh·∫•t cho m·ªói b·ªô phim|
|Year|	NƒÉm ph√°t h√†nh c·ªßa b·ªô phim|

M√¥ t·∫£ v·ªÅ d·ªØ li·ªáu c·ªßa b·∫£ng Rating:
|C·ªôt|M√¥ t·∫£|
|---|---|
|User_ID|	ƒê·ªãnh danh duy nh·∫•t cho m·ªói ng∆∞·ªùi d√πng|
|Rating|	ƒêi·ªÉm ƒë√°nh gi√° ƒë∆∞·ª£c ng∆∞·ªùi d√πng ƒë∆∞a ra cho b·ªô phim|
|Movie_ID|	ID c·ªßa b·ªô phim ƒë∆∞·ª£c ng∆∞·ªùi d√πng ƒë√°nh gi√°|

### 4. Ti·ªÅn x·ª≠ l√Ω d·ªØ li·ªáu üßπ

##### 4.1. G·ªôp 2 t·∫≠p d·ªØ li·ªáu th√†nh m·ªôt t·∫≠p duy nh·∫•t

K·∫øt h·ª£p d·ªØ li·ªáu t·ª´ hai dataframe theo c·ªôt Movie_ID

```python
combined_data = pd.merge(rating_data, movie_data, on='Movie_ID')
```

T√≠nh trung b√¨nh ƒë√°nh gi√° v√† s·ªë l∆∞·ª£ng ƒë√°nh gi√° cho m·ªói b·ªô phim

```python
combined_data = combined_data.groupby(['Movie_ID', 'Name', 'Year']).agg({'Rating': 'mean', 'User_ID': 'count'}).reset_index()
```

ƒê·ªïi t√™n c√°c c·ªôt User_ID v√† Rating

```python
# ƒê·ªïi t√™n c·ªôt User_ID th√†nh Count
combined_data = combined_data.rename(columns={'User_ID': 'Count'})

# ƒê·ªïi t√™n c·ªôt Rating th√†nh Average_Rating
combined_data = combined_data.rename(columns={'Rating': 'Average_Rating'})
```

##### 4.2. Ki·ªÉm tra s·ªë d√≤ng b·ªã tr√πng l·∫∑p v√† c√°c c·ªôt c√≥ gi√° tr·ªã thi·∫øu

Ki·ªÉm tra s·ªë d√≤ng b·ªã tr√πng l·∫∑p

```python
combined_data.isnull().sum().sort_values(ascending=False)
```

# III. Ph√¢n t√≠ch d·ªØ li·ªáu üîç

ƒê·ªÉ kh√°m ph√° v√† r√∫t ra ƒë∆∞·ª£c c√°c k·∫øt lu·∫≠n t·ª´ d·ªØ li·ªáu, ta s·∫Ω th·ª±c hi·ªán tr·∫£ l·ªùi c√°c c√¢u h·ªèi ƒë·ªÉ t·ª´ ƒë√≥ ƒë∆∞a ra c√°c ph√¢n t√≠ch v√† tr·ª±c quan h√≥a d·ªØ li·ªáu.

### 1. M·ªói User th∆∞·ªùng tham gia ƒë√°nh gi√° bao nhi√™u b·ªô phim?

ƒê·∫øm s·ªë l∆∞·ª£ng rating m√† m·ªói User ID th·ª±c hi·ªán v√† ƒë·∫∑t t√™n c·ªôt m·ªõi l√† 'Rating Per User Count'

```python
user_rating_count = rating_data['User_ID'].value_counts().reset_index()

user_rating_count.columns = ['User_ID', 'Rating Per User Count']
```

T·∫°o ra m√¥ t·∫£ cho c·ªôt Rating Per User Count

```python
user_rating_count['Rating Per User Count'].describe()
```

T·∫°o bi·ªÉu ƒë·ªì density plot cho c·ªôt Rating Per User Count

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(user_rating_count['Rating Per User Count'], color='skyblue', fill=True)
plt.title('Bi·ªÉu ƒë·ªì m·∫≠t ƒë·ªô c·ªßa `Rating Per User Count`')
plt.xlabel('Rating Per User Count')
plt.ylabel('Density')

# T√≠nh c√°c gi√° tr·ªã th·ªëng k√™
mean_value = user_rating_count['Rating Per User Count'].mean()
median_value = user_rating_count['Rating Per User Count'].median()
percentile_25 = user_rating_count['Rating Per User Count'].quantile(0.25)
percentile_75 = user_rating_count['Rating Per User Count'].quantile(0.75)
min_value = user_rating_count['Rating Per User Count'].min()
max_value = user_rating_count['Rating Per User Count'].max()

# V·∫Ω c√°c ƒë∆∞·ªùng th·∫≥ng ƒë·ª©ng cho gi√° tr·ªã trung b√¨nh, trung v·ªã v√† ph√¢n v·ªã
plt.axvline(mean_value, color='red', linestyle='--', label='Mean')
plt.axvline(median_value, color='yellow', linestyle='--', label='Median')
plt.axvline(percentile_25, color='green', linestyle='--', label='25th Percentile')
plt.axvline(percentile_75, color='blue', linestyle='--', label='75th Percentile')

# V·∫Ω c√°c ƒë∆∞·ªùng th·∫≥ng ƒë·ª©ng cho gi√° tr·ªã min v√† max
plt.axvline(min_value, color='black', linestyle='--', label='Min')
plt.axvline(max_value, color='black', linestyle='--', label='Max')

plt.legend()
plt.show()
```

- Ph√¢n ph·ªëi c·ªßa s·ªë b·ªô phim m·ªói ng∆∞·ªùi ƒë√°nh gi√° (Rating Per User Count) b·ªã l·ªách ph·∫£i, c√≥ mi·ªÅn gi√° tr·ªã t·ª´ 5 ƒë·∫øn 1344, v·ªõi gi√° tr·ªã trung b√¨nh l√† 120.85 v√† ƒë·ªô l·ªách chu·∫©n l√† 79.78. 

- Ph√¢n ph·ªëi n√†y c≈©ng cho th·∫•y r·∫±ng c√≥ m·ªôt s·ªë ng∆∞·ªùi ƒë√°nh gi√° r·∫•t nhi·ªÅu b·ªô phim, v·ªõi m·ªôt s·ªë ng∆∞·ªùi ƒë√°nh gi√° l√™n ƒë·∫øn 1344 b·ªô phim.

=> ƒêi·ªÅu n√†y cho th·∫•y r·∫±ng ph√¢n ph·ªëi c·ªßa s·ªë l∆∞·ª£ng b·ªô phim m·ªói ng∆∞·ªùi ƒë√°nh gi√° kh√¥ng ƒë·ªìng ƒë·ªÅu, v·ªõi m·ªôt s·ªë ng∆∞·ªùi ƒë√°nh gi√° √≠t b·ªô phim h∆°n so v·ªõi s·ªë ng∆∞·ªùi ƒë√°nh gi√° nhi·ªÅu b·ªô phim. ƒêi·ªÅu n√†y c√≥ th·ªÉ l√† do m·ªôt s·ªë ng∆∞·ªùi d√πng th√≠ch xem nhi·ªÅu b·ªô phim h∆°n so v·ªõi s·ªë ng∆∞·ªùi d√πng kh√°c ho·∫∑c h·ªç l√† nh·ªØng ng∆∞·ªùi l√†m c√¥ng vi·ªác li√™n quan ƒë·∫øn vi·ªác review phim.

### 10. Cho bi·∫øt Top 5 series phim c√≥ Rating trung b√¨nh cao nh·∫•t? C√°c ph·∫ßn trong Series ƒë√≥ nh·∫≠n ƒë∆∞·ª£c m·ª©c ƒë√°nh gi√° n√†o? 

T·∫°o ra danh s√°ch 5 Series phim c√≥ Rating trung b√¨nh cao nh·∫•t

```python
# L·ªçc c√°c b·ªô phim c√≥ k√≠ t·ª± ':' trong t√™n phim
season_movies = pd.DataFrame(combined_data[combined_data['Name'].str.contains(':', case=False, regex=False)])

# T·∫°o c·ªôt 'Average_Rating' t·ª´ d·ªØ li·ªáu rating ƒë√£ c√≥
season_movies['Series'] = season_movies['Name'].str.extract(r'^([^:]+):', expand=False)

# Gom nh√≥m c√°c b·ªô phim theo t√™n c·ªßa series v√† t√≠nh rating trung b√¨nh c·ªßa m·ªói series
grouped_season_movies = season_movies.groupby('Series')['Average_Rating'].mean().reset_index()

# T√≠nh s·ªë l∆∞·ª£ng ph·∫ßn phim cho m·ªói series
series_counts = season_movies.groupby('Series').size().reset_index(name='Number_of_Movies')

# T√≠nh trung b√¨nh s·ªë l∆∞·ª£t ƒë√°nh gi√° cho m·ªói series
average_counts = season_movies.groupby('Series')['Count'].mean().reset_index(name='Avg_Count').round(0)

# G·ªôp c·∫£ hai DataFrame v·ªõi DataFrame ƒë√£ c√≥
grouped_season_movies = pd.merge(grouped_season_movies, series_counts, on='Series')
grouped_season_movies = pd.merge(grouped_season_movies, average_counts, on='Series')

# L·ªçc ra c√°c series c√≥ √≠t nh·∫•t 4 phim
grouped_season_movies = grouped_season_movies[grouped_season_movies['Number_of_Movies'] >= 4]

# S·∫Øp x·∫øp theo c·ªôt 'Average_Rating' theo th·ª© t·ª± gi·∫£m d·∫ßn v√† reset index
sorted_grouped_season_movies = grouped_season_movies.sort_values(by='Average_Rating', ascending=False).reset_index(drop=True)

sorted_grouped_season_movies
```

T·∫°o bi·ªÉu ƒë·ªì th·ªÉ hi·ªán m·ª©c Rating qua c√°c ph·∫ßn c·ªßa t·ª´ng Serie phim

```python
# Danh s√°ch 5 series ƒë·∫°t ƒë∆∞·ª£c rating trung b√¨nh cao nh·∫•t
top_series = sorted_grouped_season_movies['Series'].head(5)

# L·ªçc ra c√°c b·ªô phim trong 5 series ƒë√≥ t·ª´ DataFrame combined_data
top_series_movies = combined_data[combined_data['Name'].str.contains('|'.join(top_series))].copy()

# Th√™m c·ªôt 'Series_Name' v√†o DataFrame top_series_movies
top_series_movies['Series_Name'] = top_series_movies['Name'].str.extract(r'^([^:]+):', expand=False)

# M√†u s·∫Øc cho m·ªói series
colors = ['b', 'g', 'r', 'orange', 'm']

# V·∫Ω ƒë·ªì th·ªã cho t·ª´ng series
fig, ax = plt.subplots(figsize=(12, 8))

# L∆∞u l·∫°i c√°c gi√° tr·ªã ho√†nh ƒë·ªô c·ªßa c√°c ƒë∆∞·ªùng k·∫ª l∆∞·ªõi d·ªçc
grid_lines_x = []

# T√≠nh to√°n v·ªã tr√≠ c·ªßa c√°c ƒë∆∞·ªùng k·∫ª l∆∞·ªõi d·ªçc
for i in range(len(top_series) + 1):
    grid_line_x = i - 0.5
    ax.axvline(x=grid_line_x, color='black', linewidth=0.4)
    grid_lines_x.append(grid_line_x)

# Danh s√°ch c√°c gi√° tr·ªã ho√†nh ƒë·ªô t∆∞∆°ng ·ª©ng cho 5 ƒë∆∞·ªùng trung b√¨nh
average_line_x = [grid_lines_x[i] for i in range(6)]

for i, series in enumerate(top_series):
    # L·∫•y d·ªØ li·ªáu c·ªßa t·ª´ng series
    series_data = top_series_movies[top_series_movies['Name'].str.contains(series)]
    
    # T√≠nh rating trung b√¨nh cho m·ªói b·ªô phim
    avg_ratings = series_data.groupby(['Name', 'Year'])['Average_Rating'].mean()

    # S·∫Øp x·∫øp c√°c b·ªô phim trong c√πng 1 series theo nƒÉm s·∫£n xu·∫•t
    avg_ratings = avg_ratings.sort_index(level='Year')

    # T·∫°o c√°c ƒëi·ªÉm cho m·ªói b·ªô phim
    x_values = np.linspace(i - 0.4, i + 0.4, len(avg_ratings))  # M·ªói b·ªô phim chi·∫øm 1/5 tr·ª•c Ox
    ax.scatter(x_values, avg_ratings.values, label=series, color=colors[i])

    # N·ªëi c√°c ƒëi·ªÉm v·ªõi nhau ƒë·ªÉ t·∫°o th√†nh ƒë∆∞·ªùng
    for j in range(len(avg_ratings) - 1):
        ax.plot([x_values[j], x_values[j+1]], [avg_ratings.values[j], avg_ratings.values[j+1]], color=colors[i])
    
# ƒê·∫∑t t√™n cho c√°c v√πng tr√™n tr·ª•c Ox
ax.set_xticks(np.arange(len(top_series)))
ax.set_xticklabels(top_series)

# X√≥a d·∫•u g·∫°ch ngang ·ªü m·ªói t√™n series
ax.tick_params(axis='x', length=0)

# ƒê·∫∑t ti√™u ƒë·ªÅ v√† nh√£n cho tr·ª•c y
ax.set_title('Top 5 series phim c√≥ rating trung b√¨nh cao nh·∫•t', fontsize=16)
ax.set_ylabel('Average Rating')
ax.legend()

# T√≠nh to√°n v·ªã tr√≠ c·ªßa c√°c ƒë∆∞·ªùng k·∫ª l∆∞·ªõi ngang 
for rating in np.arange(3.4, 4.7, 0.1):
    ax.axhline(y=rating, color='gray', linestyle='--', linewidth=0.4)


# Hi·ªÉn th·ªã ƒë·ªì th·ªã
plt.show()
```

### 11. C√°c ph·∫ßn c·ªßa Top 5 Series phim ƒë·∫°t ƒë∆∞·ª£c rating cao nh·∫•t c√≥ m·ª©c Rating trung b√¨nh ·ªü m·ª©c n√†o so v·ªõi m·ª©c Rating trung b√¨nh c·ªßa m·ªçi b·ªô phim t·∫°i nƒÉm ph·∫ßn phim ƒë√≥ ƒë∆∞·ª£c s·∫£n xu·∫•t?

```python
# Danh s√°ch 5 series th√†nh c√¥ng nh·∫•t
top_series = sorted_grouped_season_movies['Series'].head(5)

# L·ªçc ra c√°c b·ªô phim trong 5 series ƒë√≥ t·ª´ DataFrame combined_data
top_series_movies = combined_data[combined_data['Name'].str.contains('|'.join(top_series))].copy()

# Th√™m c·ªôt 'Series_Name' v√†o DataFrame top_series_movies
top_series_movies.loc[:, 'Series_Name'] = top_series_movies['Name'].str.extract(r'^([^:]+):', expand=False)

# Gom nh√≥m c√°c phim theo nƒÉm v√† t√≠nh trung b√¨nh rating cho m·ªói nƒÉm
average_rating_per_year = combined_data.groupby('Year')['Average_Rating'].mean()

# L·ªçc d·ªØ li·ªáu t·ª´ nƒÉm 1960 tr·ªü ƒëi
filtered_data = average_rating_per_year.loc[1960:]

colors = ['b', 'g', 'r', 'orange', 'm']

# V·∫Ω ƒë·ªì th·ªã cho t·ª´ng series v√† trung b√¨nh rating qua c√°c nƒÉm
fig, ax = plt.subplots(figsize=(12, 8))

for i, series in enumerate(top_series):
    # L·∫•y d·ªØ li·ªáu c·ªßa t·ª´ng series
    series_data = top_series_movies[top_series_movies['Name'].str.contains(series)]

    # Groupby nƒÉm v√† t√≠nh trung b√¨nh rating cho m·ªói nƒÉm
    yearly_avg_ratings = series_data.groupby('Year')['Average_Rating'].mean()
    
    # V·∫Ω ƒë·ªì th·ªã cho t·ª´ng series
    ax.plot(yearly_avg_ratings.index, yearly_avg_ratings.values, label=series, color=colors[i])

    # Th√™m ch·∫•m cho m·ªói t·∫≠p phim
    for year, rating in zip(yearly_avg_ratings.index, yearly_avg_ratings.values):
        ax.scatter(year, rating, color=colors[i])

# V·∫Ω ƒë·ªì th·ªã cho trung b√¨nh rating qua c√°c nƒÉm
ax.plot(filtered_data.index, filtered_data.values, label='Average rating', color='teal', linestyle='--')

# ƒê·∫∑t ti√™u ƒë·ªÅ v√† nh√£n cho tr·ª•c x v√† tr·ª•c y
ax.set_title('Top 5 Series phim ƒë·∫°t ƒë∆∞·ª£c rating cao nh·∫•t so v·ªõi m·ª©c rating trung b√¨nh qua c√°c nƒÉm') 
ax.set_xlabel('Year')
ax.set_ylabel('Average Rating')
ax.legend()

for rating in np.arange(3.4, 4.7, 0.1):
    ax.axhline(y=rating, color='gray', linestyle='--', linewidth=0.4)

for year in np.arange(1960, 2006, 5):
    ax.axvline(x=year, color='black', linewidth=0.4)
    
plt.show()
```
